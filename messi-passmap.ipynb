```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
# read in the data
df = pd.read_csv(r"C:\Users\nirma\Desktop\youtube-videos-main\youtube-videos-main\data\messibetis.csv")
```

```python
df
```

```python
df['x'] = df['x']*1.2
df['y'] = df['y']*0.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*0.8
```

```python
df
```

```python
fig,ax = plt.subplots(figsize=(10,5))
fig.set_facecolor('#1b3d17')
ax.patch.set_facecolor('#1b3d17')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#1b3d17', line_color='#c7d5cc', corner_arcs=True)
pitch.draw(ax=ax)

plt.gca().invert_yaxis()

for x in range (len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='#ff0000')
        plt.scatter(df['x'][x], df['y'][x], color='#ff0000')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='#000000')
        plt.scatter(df['x'][x], df['y'][x], color='#000000')
        
plt.title('Messi pass map vs Real Betis', color='#ffffff')
```
