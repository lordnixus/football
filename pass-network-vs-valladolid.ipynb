```python
import pandas as pd
from mplsoccer.pitch import Pitch
```

```python
#read in csv

df = pd.read_csv(r"C:\Users\nirma\Desktop\youtube-videos-main\youtube-videos-main\data\valladolidA.csv")
```

```python
df
```

```python
df = df[df['teamId']=='Barcelona']
```

```python
df
```

```python
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1)

passes = df[df['type']=='Pass']
successful = passes[passes['outcome']=='Successful']
```

```python
successful
```

```python
subs = df[df['type']=='SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()
```

```python
firstSub
```

```python
successful = successful[successful['minute']<firstSub]
```

```python
successful
```

```python
pas = pd.to_numeric(successful['passer'], downcast='integer')
rec = pd.to_numeric(successful['recipient'], downcast='integer')

successful['passer'] = pas
successful['recipient'] = rec
```

```python
successful
```

```python
avg_locations = successful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
avg_locations.columns = ['x', 'y', 'count']
```

```python
avg_locations
```

```python
pass_between = successful.groupby(['passer', 'recipient']).id.count().reset_index()
pass_between.rename({'id':'pass_count'}, axis='columns', inplace = True)

pass_between = pass_between.merge(avg_locations, left_on = 'passer', right_index = True)
pass_between = pass_between.merge(avg_locations, left_on = 'recipient', right_index = True, suffixes = ['', '_end'])
```

```python
pass_between
```

```python
pass_between = pass_between[pass_between['pass_count'] > 3]
```

```python
pass_between
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc', corner_arcs=True)
fig, ax = pitch.draw(figsize=(11, 6), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

arrows = pitch.arrows(1.2 * pass_between.x, 0.8 * pass_between.y, 1.2 * pass_between.x_end, 0.8 * pass_between.y_end, ax=ax, color = '#ff0000',
                     width = 2.5, headwidth = 3, zorder = 5, alpha = 0.5)

nodes = pitch.scatter(1.2 * avg_locations.x, 0.8 * avg_locations.y,
                     s = 300, color = '#d3d3d3', edgecolors = 'black', linewidth = 2.5, alpha = 1, zorder = 1, ax=ax)
```
